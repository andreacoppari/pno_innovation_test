### Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "../data/dataset_10k.jsonl"
df = pd.read_json(path, lines=True)
df.shape, df.columns.tolist()[:10]
```

```python
df.head(10)[["id","title","categories","update_date","doi"]]
```

```python
df["abstract_len"] = df["abstract"].fillna("").str.len()
sns.histplot(df["abstract_len"], bins=100, color="#6366f1")
plt.title("Abstract length distribution"); plt.xlabel("Characters"); plt.ylabel("Frequency")
plt.tight_layout()
```

P.S.: This is why I chose 800 chars as Text Chunking size, so to keep the average abstract intact, thus not losing information. Text splitting increases accuracy for long texts.
