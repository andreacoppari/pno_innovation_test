```python
%%bash
curl -L -o ../data/arxiv.zip https://www.kaggle.com/api/v1/datasets/download/Cornell-University/arxiv
unzip ../data/arxiv.zip -d ../data/
mv ../data/arxiv-metadata-oai-snapshot.json ../data/dataset.jsonl
```

```python
import json
from tqdm import tqdm

dataset_10k = []
dataset_10k_jsonl = []

with open("../data/dataset.jsonl", "r") as fp:
    for i, line in enumerate(tqdm(fp, desc="Loading dataset entries", total=10000)):
        if i<10000:
            dataset_10k.append(json.loads(line))
            dataset_10k_jsonl.append(line)
        else:
            break

with open("../data/dataset_10k.json", "w") as fout:
    json.dump(dataset_10k, fout, indent=4)

with open("../data/dataset_10k.jsonl", "w") as fout:
    for line in dataset_10k_jsonl:
        fout.write(f'{line}')
```
